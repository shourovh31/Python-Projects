```python
#preparing dataset for Random Forest Classification Model
```

```python
loan.info()
```

```python
#since the purpose column as categorical, it is necessary to transform them using DUMMY VARIABLES, so that scikuit learn understand them
```

```python
#creating a list containing a string 'purpose'
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loan, columns=cat_feats, drop_first = True)
```

```python
final_data.head()
```

```python
#Training the dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X =final_data.drop('not.fully.paid', axis = 1)

y = final_data['not.fully.paid']


X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
#Now training the decsion tree model
```

```python
#import decision tree algorithm (DecisiontreeClassifier)

from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix( y_test, predictions))
```

```python
#Now train in random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc =  RandomForestClassifier(n_estimators = 500)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_pred))
print('\n')
print(classification_report(y_test, rfc_pred))
```
