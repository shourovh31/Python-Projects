```python
#using lambda function
#Seasons for race

#Is summer slower than winter?
#summer 6 - 8
#spring 3-5
#winter 12-2

#split between decimals
```

```python
df3['Race Month'] = df3['Race Day'].str.split('.').str.get(1).astype(int)
```

```python
df3
```

```python
df3['Race Season'] = df3['Race Month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x >2 else 'Winter')
```
