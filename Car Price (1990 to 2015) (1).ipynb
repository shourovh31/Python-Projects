```python
#data Cleaning
```

```python
#Renaming Columns
df = df.rename(columns = { 'year' : 'Year',
                          'make' : 'Brand Name',
                          'trim' : 'Version',
                          'body' : 'Body',
                          'transmission': 'Transmission',
                          'vin': 'Vin',
                          'state': 'State',
                          'condition' : 'Condition',
                          'color': 'Color',
                          'odometer': 'Odometer',
                          'interior' : 'Interior',
                          'seller' : 'Seller',
                          'mmr' : 'Manheim Market',
                          'sellingprice' : 'Price',
                          'saledate' : 'Sale Date'})
```

```python
df.head(2)
```

```python
#Capitalize row values from columns
```

```python
capitaliz_columns = ['Transmission','Color', 'Interior']
```

```python
for col in capitaliz_columns:
    df[col] = df[col].str.capitalize()
```

```python
df.head(2)
```

```python
#Upper State colunm
```

```python
df['State'] = df['State'].str.upper()
```

```python
#drop Unneccessary Columns
```

```python
df.drop(['Vin','Odometer','Condition'], axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
#Capitalizing each words in Seller Column
```

```python
df['Seller']= df['Seller'].str.title()
```

```python
df.head(2)
```

```python
#Extracting Day name from the Sale date column and creating a new column for day
```

```python
df['Day'], df['Sale Date'] = df['Sale Date'].str[:3], df['Sale Date'].str[3:]
```

```python
#Adding Month Column
df['Month'] = df['Sale Date'].str.extract(r'(\w+)')[0]
```

```python
# Removing the time information from the 'Sale Date' column
df['Sale Date'] = df['Sale Date'].str.replace(r'\s+\d{2}:\d{2}:\d{2}.*$', '', regex=True)
```

```python
df.head(2)
```

```python
df['Day'].value_counts()
```

```python
#Deleting Null Values from df
```

```python
df.dropna(inplace = True)
```

```python
#Checking null values after deleting null values
df.isnull().sum()
```
